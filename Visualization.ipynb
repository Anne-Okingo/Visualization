```python
# Exercise 4: Matplotlib 2

import matplotlib.pyplot as plt

# Data
left_data = [5, 7, 11, 13, 17]
right_data = [0.1, 0.2, 0.4, 0.8, -1.6]
x_axis = [0.0, 1.0, 2.0, 3.0, 4.0]

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot left data on primary y-axis (black)
ax1.plot(x_axis, left_data, color='black')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('Big')

# Create secondary y-axis
ax2 = ax1.twinx()

# Plot right data on secondary y-axis (red)
ax2.plot(x_axis, right_data, color='red')
ax2.set_ylabel('Small', color = 'red')
ax2.tick_params(axis='y', labelcolor='red')


# Set title
plt.title('Twin axes')

# Adjust layout to prevent label cutoff
# fig.tight_layout()

# Display the plot
plt.show()
```

```python
# Exercise 5: Matplotlib subplots
import matplotlib.pyplot as plt

# Create figure with 2 rows and 3 columns of subplots
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# Adjust space between plots
plt.subplots_adjust(hspace=0.5, wspace=0.5)

# Flatten the axes array for easier iteration
axes_flat = axes.flatten()

# Loop through each subplot
for i in range(6):
    ax = axes_flat[i]
    
    # Set title
    ax.set_title(f'Title {i+1}', fontsize=12)
    
    # Set axis limits
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    
    # Add text centered at (0.5, 0.5)
    ax.text(0.5, 0.5, f'(2, 3, {i+1})', 
            fontsize=16,
            ha='center',  # horizontal alignment
            va='center')  # vertical alignment

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Generate data
y0 = np.random.randn(50)
y1 = np.random.randn(50) + 1  # shift mean by +1
y2 = np.random.randn(50) + 2  # shift mean by +2

# Create DataFrame in long format
df = pd.DataFrame({
    'Sample A': y0,
    'Sample B': y1,
    'Sample C': y2
})

# Melt DataFrame to long form
df_melt = df.melt(var_name='Sample', value_name='Value')

# Plot
fig = px.box(df_melt, x='Sample', y='Value', color='Sample',
             title='Box plot comparison of three samples')

fig.show()
```
